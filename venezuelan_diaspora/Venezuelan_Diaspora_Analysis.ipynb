```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of people who answered "Yes" and "No" to the migration question
migration_counts = df['Did you migrate from Venezuela?'].value_counts().reset_index()
migration_counts.columns = ['Migration Status', 'Count']

# Calculate the percentage of each category
total = migration_counts['Count'].sum()
migration_counts['Percentage'] = (migration_counts['Count'] / total) * 100

# Create the bar plot
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(data=migration_counts, x='Migration Status', y='Count', palette='viridis')
plt.title('Number of People Migrating from Venezuela')
plt.xlabel('Migration Status')
plt.ylabel('Number of People')

# Add labels to the bars
for index, row in migration_counts.iterrows():
    bar_plot.text(row.name, row['Count'], f'{row["Count"]} ({row["Percentage"]:.1f}%)', color='black', ha="center")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter the DataFrame to include only rows where people migrated from Venezuela
venezuela_migrants = df[df['Did you migrate from Venezuela?'] == 'Yes']

# Filter out rows where the ancestral information is missing or undefined
venezuela_migrants_with_ancestors = venezuela_migrants.dropna(subset=['Where did your family come from?'])

# Create a dictionary mapping countries to continents
country_to_continent = {
    'North America': ['United States', 'Canada'],
    'South America': ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela'],
    'Europe': ['Portugal', 'Italy', 'Spain', 'France', 'Austria', 'Germany', 'United Kingdom', 'Poland', 'Netherlands'],
    'Asia': ['Lebanon'],
    'Other': ['Trinidad & Tobago', 'Cuba']
}

# Define a function to check if the country is in any continent
def is_country_in_continent(country):
    for continent, countries in country_to_continent.items():
        if country in countries:
            return True
    return False

# Apply the filter function to the DataFrame
descendants_by_continent = venezuela_migrants_with_ancestors[venezuela_migrants_with_ancestors['Where did your family come from?'].apply(is_country_in_continent)]

# Filter out rows where the ancestral continent is "Other"
descendants_by_continent = descendants_by_continent[descendants_by_continent['Where did your family come from?'] != 'Trinidad & Tobago']
descendants_by_continent = descendants_by_continent[descendants_by_continent['Where did your family come from?'] != 'Cuba']

# Map the countries to continents
descendants_by_continent['Continent'] = descendants_by_continent['Where did your family come from?'].map({country: continent for continent, countries in country_to_continent.items() for country in countries})

# Calculate the percentage of migrants for each continent
continent_percentage_ancestors = descendants_by_continent['Continent'].value_counts(normalize=True).mul(100).reset_index()
continent_percentage_ancestors.columns = ['Continent', 'Percentage']

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot_ancestors = sns.barplot(data=continent_percentage_ancestors, x='Continent', y='Percentage', palette='rocket')
plt.title('Percentage of Venezuelan Migrants by Ancestral Continent')
plt.xlabel('Ancestral Continent')
plt.ylabel('Percentage')

# Add labels to the bars
for index, row in continent_percentage_ancestors.iterrows():
    bar_plot_ancestors.text(row.name, row['Percentage'], f'{row["Percentage"]:.1f}%', color='black', ha="center")

# Set y-axis limits from 0 to 100
plt.ylim(0, 100)

plt.tight_layout()
plt.show()
```
